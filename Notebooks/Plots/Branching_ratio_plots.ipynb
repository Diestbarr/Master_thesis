```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import normal
```

```python
Observables = pd.read_pickle("Observables")

Observables
```

```python
Branchings = Observables[Observables["Name"].str.contains("Br")]

Branchings
```

```python
B_plus_to_K_plus = pd.DataFrame(Branchings[Branchings["Name"]=="Br(B+-->K+mumu)"]) #Extracts the branching ratios of the B+ -> K+ process

B_plus_to_K_plus
```

```python
B_zero_to_K_zero = pd.DataFrame(Branchings[Branchings["Name"]=="Br(B0-->K0mumu)"]) #Extracts the branching ratios of the B0 -> K0 process

B_zero_to_K_zero
```

```python
B_zero_to_K_star = pd.DataFrame(Branchings[Branchings["Name"]=="Br(B0-->K0*mumu)"]) #Extracts the branching ratios of the B0 -> K* process

B_zero_to_K_star
```

```python
B_plus_to_K_star = pd.DataFrame(Branchings[Branchings["Name"]=="Br(B+-->K+*mumu)"]) #Extracts the branching ratios of the B+ -> K+* process

B_plus_to_K_star
```

Let's add the central value of the bin and the length from the center to the end of the interval

```python
B_plus_to_K_plus["CM"] = [0.54,1.55,2.5,3.5,4.5,5.5,6.5,7.5,18.5] #Class mark of each bin
B_plus_to_K_plus["Side"] = [0.44,0.45,0.5,0.5,0.5,0.5,0.5,0.5,3.5] #Side length of each bin

B_plus_to_K_plus
```

```python
B_zero_to_K_zero["CM"] = [1.05,3,5,7,18.5] #Class mark of each bin
B_zero_to_K_zero["Side"] = [0.95,1,1,1,3.5] #Side length of each bin

B_zero_to_K_zero
```

```python
B_zero_to_K_star["CM"] = [0.54,1.8,3.25,5,7,17] #Class mark of each bin
B_zero_to_K_star["Side"] = [0.44,0.7,0.75,1,1,2] #Side length of each bin

B_zero_to_K_star
```

```python
B_plus_to_K_star["CM"] = [1.05,3,5,7,17] #Class mark of each bin
B_plus_to_K_star["Side"] = [0.95,1,1,1,2] #Side length of each bin

B_plus_to_K_star
```

Let's plot the comparisons

```python
x = B_plus_to_K_plus.CM
Side = B_plus_to_K_plus.Side
Th = B_plus_to_K_plus.Th_Pred
Exp = B_plus_to_K_plus.Exp_Value
sigma_Th = B_plus_to_K_plus.Th_Err
sigma_Exp = B_plus_to_K_plus.Exp_Err


fig,(ax,ax2) = plt.subplots(1,2,sharey=True, facecolor='w')

ax.errorbar(x,Th,xerr=Side, yerr=sigma_Th, linestyle='', capsize=3, label="Theory")
ax.errorbar(x,Exp,xerr=Side, yerr=sigma_Exp, linestyle='', capsize=3, label="Experiment")
ax.legend(loc="upper left")


ax.set_ylabel("Branching ratio", fontsize=16)
ax.set_xlim(-0.5,8.5)


ax2.errorbar(x,Th,xerr=Side, yerr=sigma_Th, linestyle='', capsize=3, label="Theory")
ax2.errorbar(x,Exp,xerr=Side, yerr=sigma_Exp, linestyle='', capsize=3, label="Experiment")

ax2.set_xlim(14.5,22.5)


#Sets the broken x axis

ax.spines.right.set_visible(False)
ax2.spines.left.set_visible(False)
ax.yaxis.tick_left()
ax.tick_params(labeltop=False)  # don't put tick labels at the top
ax2.yaxis.tick_right()

fig.tight_layout()

d = .7  # proportion of vertical to horizontal extent of the slanted line
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax.plot([1, 1], [0, 1], transform=ax.transAxes, **kwargs)
ax2.plot([0, 0], [1, 0], transform=ax2.transAxes, **kwargs)





fig.suptitle(r'$B^+ \to \: K^+ \mu^+ \mu^-$', x=0.55, y=1.05, va="top", fontsize=16)
fig.supxlabel(r'$q(GeV^2)$', fontsize=16, x=0.55, y=-0.05)



#plt.savefig("B_plus_to_K_plus.jpg", bbox_inches='tight', dpi=300)
plt.show()
```

Lets compute the chi^2 difference

```python
Chi = np.array((B_plus_to_K_plus.Diff/B_plus_to_K_plus["Joint_Err"])**2) #Calculates the contribution for each piece of the chi^2
bins = np.array(B_plus_to_K_plus.Bin) #Stores the bins in an array

sum_chi = np.round(np.sum(Chi),2)

plt.bar(bins,Chi)
plt.title(r'$B^+ \to \: K^+ \mu^+ \mu^-$', fontsize=16)
plt.xlabel(r'$q(GeV^2)$', fontsize=16)
plt.ylabel(r'$\chi^2$', fontsize=16)

plt.text(x=6, y=1.2, s=r'$\chi^2=$%s'%sum_chi)

#plt.savefig("Chi_B_plus_to_K_plus.jpg",bbox_inches="tight", dpi=300)

plt.show()
```

```python
Chi = np.array((B_zero_to_K_zero.Diff/B_zero_to_K_zero["Joint_Err"])**2) #Calculates the contribution for each piece of the chi^2
bins = np.array(B_zero_to_K_zero.Bin) #Stores the bins in an array

sum_chi = np.round(np.sum(Chi),2)

plt.bar(bins,Chi)
plt.title(r'$B^0 \to \: K^0 \: \mu^+ \mu^-$', fontsize=16)
plt.xlabel(r'$q(GeV^2)$', fontsize=16)
plt.ylabel(r'$\chi^2$', fontsize=16)

plt.text(x=2, y=2.5, s=r'$\chi^2=$%s'%sum_chi)

#plt.savefig("Chi_B_zero_to_K_zero.jpg",bbox_inches="tight", dpi=300)

plt.show()
```

```python
Chi = np.array((B_plus_to_K_star.Diff/B_plus_to_K_star["Joint_Err"])**2) #Calculates the contribution for each piece of the chi^2
bins = np.array(B_plus_to_K_star.Bin) #Stores the bins in an array

sum_chi = np.round(np.sum(Chi),2)

plt.bar(bins,Chi)
plt.title(r'$B^+ \to \: K^{+*} \: \mu^+ \mu^-$', fontsize=16)
plt.xlabel(r'$q(GeV^2)$', fontsize=16)
plt.ylabel(r'$\chi^2$', fontsize=16)

plt.text(x=2, y=6, s=r'$\chi^2=$%s'%sum_chi)

plt.savefig("Chi_B_plus_to_K_star.jpg",bbox_inches="tight", dpi=300)

plt.show()
```
